```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as wdg  # Using the ipython notebook widgets
import numpy as np
from IPython import display

# Create a random image
# a = np.random.poisson(size=(15,15))
fig = plt.figure(figsize=(6,6))
# plt.imshow(a)
theta = np.array(range(100))/99*np.pi*2
plt.plot(np.cos(theta),np.sin(theta), '-k')
plt.xlim([-1,1])
plt.ylim([-1,1])
# Create and display textarea widget
txt = wdg.Textarea(
    value='',
    placeholder='',
    description='event:',
    disabled=False
)
# display(txt)

xlist = []
ylist = []
# Define a callback function that will update the textarea
def onclick(event):
    txt.value = str(event)  # Dynamically update the text box above
    xlist.append(event.xdata)
    ylist.append(event.ydata)
    plt.scatter(xlist,ylist,color='grey')
    display.display(plt.gcf())
    display.clear_output(wait=True)


    
# Create an hard reference to the callback not to be cleared by the garbage collector
ka = fig.canvas.mpl_connect('button_press_event', onclick)
```

```python
xlist, ylist = np.array(xlist), np.array(ylist)
rlist = np.sqrt(xlist**2 +  ylist**2)

accept = 0
reject = 0
for i in range(len(rlist)):
    if rlist[i]<1.0:
        accept += 1
    else: 
        reject += 1
print(accept/len(rlist)*4)
        
        
```
